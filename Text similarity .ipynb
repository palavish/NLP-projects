```python
def remove_pattern(txt,pattern):
    r=re.findall(pattern,txt)
    for word in r:
        txt=re.sub(word, "",txt)
    return txt
```

```python
df['clean_title']=np.vectorize(remove_pattern)(df['Title'], "@[\w |]*")
df
```

```python
df['clean_title']=df['clean_title'].str.replace("[^a-zA-Z#]"," ")
df.head()
```

```python
tokenized_title=df['clean_title'].apply(lambda x: x.split())
tokenized_title.head()
```

```python
from nltk.stem.porter import PorterStemmer as ps
stemmer =ps()

tokenized_title=tokenized_title.apply(lambda sen:[stemmer.stem(word) for word in sen])
tokenized_title.head()
```

```python
for i in range(len(tokenized_title)):
    tokenized_title[i]=" ".join(tokenized_title[i])
df['clean_title']=tokenized_title
df.head()
```

```python
score=[]
for i in range(len(df['clean_title'])):
    s1=all_docs[i]
    for j in range(len(df['clean_title'])):
        sc=s1.similarity(all_docs[j])
    df['s_score'][i]=sc
    
```
